```python
def max_of_two(a,b):
    return a if a > b else b

print(max_of_two(2,5))
```

#A3

```python
def is_prime(a):
    for i in range (2, int(a**0.5) + 1):
        if a % i == 0:
            return False
    return True


print(is_prime(101))
```

#A4

```python
def is_perfect_number(a):
    divisor_sum = 1
    for i in range(2, int(a**0.5)+1):
        if a % i == 0:
            divisor_sum += i
            paired_divisor = a / i
            if paired_divisor != a:
                divisor_sum += paired_divisor 
    return True if divisor_sum == a else False

print(is_perfect_number(29))
```

#A5

```python
def find_num(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1


a = [1,2,3,6,2]
find_num(a, 6)
```

#A6

```python
def factorial(a):
    if a == 1:
        return 1
    return a * factorial(a-1)
```

#A7

```python
def calculator(a, b, operator):
    if operator == "+":
        result = a + b
    if operator == "-":
        result = a - b
    if operator == "*":
        result = a * b 
    if operator == "/":
        result = a / b
    return f"{result:.2f}"

print(calculator(5,2,"/"))
```

#A8

```python
def hamming(a, b):
    a, b = bin(a), bin(b)
    result = 0
    for i in range(2, min(len(a), len(b))):
        if a[i] != b[i]:
            result +=1
    result += abs(len(a)-len(b))

    return result

print(hamming(15,8))
    
    
```

```python
def sum_digits(a):
    if len(str(a)) == 1:
        return a        
    temp = a % (10**(len(str(a))-1)) #bỏ đi chữ số ngoài cùng bên trái
    return a // (10**(len(str(a))-1)) + sum_digits(temp) # Tổng giữa số có độ dài n và số có độ dài n-1 (sau khi bỏ đi 1 chữ số ở bước trên)

print(sum_digits(652))
```

```python
def isomorphism(w1, w2):
    if len(w1) != len(w2):
        return False
        
    for i in range(len(w1)):
        indices = [] 
        for k, char in enumerate(w1):
            if char == w1[i]:
                indices.append(k) #Lấy index của toàn bộ vị trí từng kí tự trong w1
        for j in indices:
            if w2[i] != w2[j]: 
                return False #Dùng index bên trên để tạo pattern so sánh các kí tự trong w2
    return True
            
            

a = "paper"
b = "title"

print(isomorphism(a, b))


    
```
