```python
def max_of_two(a,b):
    return a if a > b else b

print(max_of_two(2,5))
```

```python
def swap(a, b):
    a, b = b, a
    return a, b

print(swap(5,2))
```

```python
def is_prime(a):
    for i in range (2, int(a**0.5) + 1):
        if a % i == 0:
            print("haha")
            return False
    return True


print(is_prime(4))
```
